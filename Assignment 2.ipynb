5. Filtering Data:
   - Select all data related to a continent of your choice (e.g., Africa).
   - Filter out countries with a population less than 1 million. 

```python
def convert_values(x):
    if isinstance(x, str):
        if 'k' in x:
            return float(x.replace('k', '').strip()) * 1000
        elif 'M' in x:
            return int(float(x.replace('M', '').strip()) * 1000000)
        elif 'B' in x:
            return int(float(x.replace('B', '').strip()) * 1000000000)
    return x 
all_data['Population'] = all_data['Population'].apply(convert_values)
print(all_data['Population'])
#In the data, the population variable was coded as a string and the values ​​of thousand, million and billion were represented by k, m, b. 
#Therefore, I write a function to convert the values ​​to integers by multiplying k by 1000, m by 1000000 and b by 1000000000.
#Then I check if it works. (I can do it for other variables too, but since I don't need them now I let it be.)
```

```python
 europe_countries = ['Albania', 'Andorra', 'Armenia', 'Austria', 'Azerbaijan', 'Belarus', 'Belgium', 
                    'Bosnia and Herzegovina', 'Bulgaria', 'Croatia', 'Cyprus', 'Czech Republic', 'Denmark', 
                    'Estonia', 'Finland', 'France', 'Georgia', 'Germany', 'Greece', 'Hungary', 'Iceland', 
                    'Ireland', 'Italy', 'Kazakhstan', 'Kosovo', 'Latvia', 'Liechtenstein', 'Lithuania', 
                    'Luxembourg', 'Malta', 'Moldova', 'Monaco', 'Montenegro', 'Netherlands', 'North Macedonia', 
                    'Norway', 'Poland', 'Portugal', 'Romania', 'Russia', 'San Marino', 'Serbia', 'Slovakia', 
                    'Slovenia', 'Spain', 'Sweden', 'Switzerland', 'Turkey', 'Ukraine', 'United Kingdom', 'Vatican City']
#I get a list of countries in Europe with the help of ChatGPT and define a function to search them in all data.
def filtering (all_data):
    filter_europe = [] #I create an empty list to add countries with the qualifications I want.
    for index, row in all_data.iterrows():
        if row['country'] in europe_countries and int(row['Population']) > 1000000: 
#I check whether the country is in my europe_countries list and have population more than 1000000.
            filter_europe.append(row)
    return filter_europe
#If a country meet my criteria I put the row on filter_europe list. 
filter_europe = filtering(all_data)
filter_europe_df = pd.DataFrame(filter_europe)
display(filter_europe_df)
#Lastly, I convert the data that I filtered and assigned to the list into a pandas dataframe and check it.
```

6. Data Organization:
   - Sort the dataset based on life expectancy in ascending order.
   - Create a new column that categorizes countries into 'High', 'Medium', or 'Low' 
population based on defined thresholds.

```python
sort_data = all_data.sort_values(by='Life_expectancy')
display(sort_data)
#I sort the data according to life expectancy. Since it sorts descending by default, I don't need to give any extra commands.
```

```python
all_data['Population'] = pd.to_numeric(all_data['Population']) #Normally I don't need this but something happened and Population was seen as str, so I convert it to 
def grouping (group):
    min_pop = group['Population'].min()
    max_pop = group['Population'].max()
    threshold_value = (max_pop - min_pop) / 3
    threshold_1 = min_pop + threshold_value
    threshold_2 = threshold_1 + threshold_value
#I develop a simple threshold calculation method because I don't want the values ​​to be random.
#By defining a function, I make a calculation that will evaluate each country's population by year.
#I take the difference between them and divide by 3, which gives me a threshold range. 
#Then I add this threshold_value to the minimum population value to determine the treshold_1 and add it to treshold_1 to find the treshold_2.
   
    def categorizing (group):
            if group <= threshold_1:
                return 'Low'
            elif group <= threshold_2:
                return 'Medium'
            else:
                return 'High'
    group['Population'] = group['Population'].apply(categorizing)
    return group
 #I write another function to evaluate the population of each country according to the thresholds I set.
#If the population ​​of countries are equal to or less than threshold_1, I mark them as low.
#If they are equal to or less than threshold_, I mark them as medium.
#In any other condition I mark them as high.
all_data_new = all_data.groupby('year').apply(grouping)
#Since I want to make separate evaluations for each year, I first group my data by year.
display(all_data_new) 
```
